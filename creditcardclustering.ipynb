CREDIT CARD CUSTOMER DATA CLUSTURING (using K-means clustering - N.Nandhini)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/Credit Card Customer Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(df)
lables=kmeans.labels_
print(lables)
```

```python
plt.scatter(df.iloc[:, 2], df.iloc[:, 3], c=lables)
# plt.scatter(
#     df.iloc[:, 0],
#     df.iloc[:, 1],
#     marker='x',
#     s=200
# )

plt.xlabel(df.columns[2])
plt.ylabel(df.columns[3])
plt.title("Credit card customer data")

plt.show()
```

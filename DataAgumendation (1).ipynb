```python

import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
```

```python
DATA_DIR = '/content/drive/MyDrive/Colab Notebooks/Source/DataAug'
```

```python
AUGMENT_FACTOR = 20
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
print(f"Scanning {DATA_DIR}...")
```

```python

total_generated = 0

for root, dirs, files in os.walk(DATA_DIR):
    for filename in files:
        # Check for image files only
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):

            img_path = os.path.join(root, filename)

            try:
                # Load the image
                img = load_img(img_path)
                x = img_to_array(img)
                x = x.reshape((1,) + x.shape)  # Reshape to (1, w, h, c)

                # Generate augmented images
                i = 0
                for batch in datagen.flow(x, batch_size=1,
                                          save_to_dir=root,      # Save in the SAME class folder
                                          save_prefix='aug',     # File name prefix
                                          save_format='jpeg'):
                    i += 1
                    total_generated += 1
                    if i >= AUGMENT_FACTOR:
                        break  # Stop after generating the desired amount
            except Exception as e:
                print(f"Error processing {filename}: {e}")

print(f"Done! Created {total_generated} new images.")
```
